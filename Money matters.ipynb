```python
# importing time module to make use of the sleep function in it.

import time

# Declaring empty lists to store data as required

needs = [] # stores the names of needs
wants = [] # stores the names of wants
pr1 = [] # stores the priority of needs
pr2 = [] # stores the priority of wants
p1 = [] # stores the price of each needs
p2 = [] # stores the price of each wants



# Defining sort function to sort all the lists on the basis of the priority

def sort(a_list, l1, l2):
    
    # Using Bubble sort technique to sort
    
    l = len(a_list)
    for i in range(l):
        for j in range(0, l-i-1):
            if(a_list[j] < a_list[j+1]):
                a_list[j],a_list[j+1] = a_list[j+1],a_list[j] # Swaping the lists
                l1[j], l1[j+1] = l1[j+1], l1[j]
                l2[j], l2[j+1] = l2[j+1], l2[j]
            else:
                pass
    return ""

# Defining calc() function which calculates the budget based on needs and wants

def calc(diff,sav):
    sum1 = 0.0
    diff_1 = diff
    
    # displaying the needs and wants in a sorted format
    
    print("\n So As per The given data your Monthly Budget will be as followed :")
    print("\n :::::::::::Your Need's List::::::::::: ")
    print("\n | Name |         |Priority|        |Price|   \n")
    for i in range(0, len(needs)):
        print("\n ",needs[i],"\t\t", pr1[i],"\t\t", p1[i],"\n")
    print("\n\n :::::::::::Your Want's List::::::::::: ")
    print("\n | Name |         |Priority|        |Price|   \n")
    for j in range(0, len(wants)):
        print("\n ",wants[j],"\t\t", pr2[j],"\t\t", p2[j])
    
    # Calculating Budget by adding all items which can sustain user's budget
    
    i=0
    while(diff>0 and i<len(needs)):
        if((diff-p1[i])>0):
            diff=diff-p1[i]
            print("\n Amount left after fulfilling your ",needs[i]," need is: ",diff)
            i=i+1
        else:
            print("\n So Sorry but Your ", needs[i], " need can't be satisfied as it exceeds your Budget")
            i+=1
    j=0
    while(diff>0 and j<len(wants)):
        if((diff-p2[j])>0):
            diff=diff-p2[j]
            print("\n Amount left after fulfilling your ",wants[j]," want is: ",diff)
            j=j+1
        else:
            print("\n So Sorry but Your ", wants[j], " want can't be satisfied as it exceeds your Budget")
            j+=1
    
    # Printing the final results
    
    print("\n\n So, The total Amount You are spending = ", (diff_1 - diff) )
    print("\n Total Amount that you are saving = ",(sav + diff))
    
    # Asking the user to re-run if he didn't like the results
    
    print("\n If not satisfied with the results then : ")
    print("\n Enter 1 To Re-run The program")
    print("\n Enter 2 To Exit")
    choice = int(input(" Enter your choice = "))
    if(choice == 1):
        print(disp())
    elif(choice == 2):
        print("\n  Thank You For being With Us :)")
    return ""

# Defining need function which builds a complete list of user's needs

def need(diff,sav):
    print("\n\n*****Please enter the list of your NEEDS acoording to your priority*****")
    print("\n Don't forget to provide the price and priority value ")
    print("\n The priority will be marked on a scale of 1-10, 10 means it has the highest priority")
    size = int(input("\n\n Enter the number of needs you have = "))
    s = ""
    for i in range(1, size + 1):
        if i == 1:
            s = s + str(i) + "st "
        elif i == 2:
            s = s + str(i) + "nd "
        elif i == 3:
            s = s + str(i) + "rd "
        else:
            s = s + str(i) + "th "
        c1 = input("Enter your "+ s +"need = ")
        c2 = int(input("Enter its priority value = "))
        c3 = float(input("Enter its Price = "))
        s = ""
        needs.append(c1) 
        pr1.append(c2)
        p1.append(c3)
    
    sort(pr1, needs, p1)
    
    # Checking whether wants list is filled or not
    
    if(len(wants)==0):
        print("\n You have not entered Your List Of Wants ")
        print("\n Enter 1 to Enter The list of wants ")
        print("\n\t\t OR ")
        print("\n Enter 2 To Continue without any Wants")
        ch = int(input("\n Enter Your Choice = "))
        if(ch == 1):
            print("\n Redirecting You to Wants Section.....")
            want(diff,sav)
        elif(ch == 2):
            calc(diff,sav)
    else:
        print("\n You have provided all your inputs!")
        print("\n Calculating your finalized Budget")
        calc(diff,sav)
    return ""    

# defining want() function to build a list of user's wants 
    
def want(diff,sav):
    print("\n\n*****Please enter the list of your WANTS acoording to your priority*****")
    print("\n Don't forget to provide the price and priority value ")
    print("\n The priority will be marked on a scale of 1-10, 10 means it has the highest priority")
    size = int(input("\n\n Enter the number of wants you have = "))
    s = ""
    for i in range(1, size + 1):
        if i == 1:
            s = s + str(i) + "st "
        elif i == 2:
            s = s + str(i) + "nd "
        elif i == 3:
            s = s + str(i) + "rd "
        else:
            s = s + str(i) + "th "
        c1 = input("Enter your "+ s +" want = ")
        c2 = int(input("Enter its priority value = "))
        c3 = float(input("Enter its Price = "))
        s = ""
        wants.append(c1) 
        pr2.append(c2)
        p2.append(c3)
    
    sort(pr2, wants, p2)
    
    # Checking whether needs list is filled or not
    
    if(len(needs)==0):
        print("\n You have not entered Your List Of Needs ")
        print("\n Enter 1 to Enter The list of Needs ")
        print("\n\t\t OR ")
        print("\n Enter 2 To Continue without any Needs")
        ch = int(input("\n Enter Your Choice = "))
        if(ch == 1):
            print("\n Redirecting You to Needs Section.....")
            need(diff,sav)
        elif(ch == 2):
            calc(diff,sav)
    else:
        print("\n You have provided all your inputs!")
        print("\n Calculating your finalized Budget")
        calc(diff,sav)
    return ""

# define disp() function which takes budget and savings as inputs and connects all the other functions

def disp():
    s = "\n ******** Welcome to the Monthly Budget Planner ********"
    for i in s:
        print(i, end=" ")
        time.sleep(0.03) # Adding a pause 3 milli seconds
    budget = float(input("\n\nPlease enter your Monthly budget = "))
    saving = float(input("\nPlease enter the amount you want to save = "))
    diff = budget - saving
    
    # checking if saving is greater than budget then re-enter
    
    if(saving > budget):
        print("\nYou can't save more than your budget .... Please provide your inputs again!!! ")
        disp()
    else:      
        print("\nPress 1 To enter the list of needs ")
        print("\nPress 2 To enter the list of wants ")
        c1 = int(input("Enter your choice = "))
        if c1 == 1:
            need(diff,saving)
        elif c1 == 2:
            want(diff,saving)
    return ""

# calling to the disp() function to initiate the program from beginning

print(disp())
```
